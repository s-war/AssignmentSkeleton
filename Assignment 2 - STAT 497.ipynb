```python
#Sadia Waraich - 40158487
#Assignment 2

import matplotlib.pyplot as plt

def draw_tennis_court(filename="tennis_court.png"):
    # Create a figure and the axis
    fig, ax = plt.subplots(figsize=(10, 5))

    # Write the court dimensions with their value
    court_length_ft = 78
    court_width_ft = 36
    service_lane_length_ft = 18  
    service_lane_width_ft = 27  # Make sure that the adjusted service lane width matches the court width
    net_height_ft = 3.5  # The standard net height
    middle_line_length_ft = 42
    double_alley_width_ft = 4.5

    # Name the X-axis and Y-axis
    ax.set_xlabel('Length (feet)')
    ax.set_ylabel('Width (feet)')
    
    # Title
    ax.set_title('Tennis Court')
    
    # Set the limits for the axis
    ax.set_xlim(0, court_length_ft)
    ax.set_ylim(0, court_width_ft)

    # Draw the court's outer boundary
    outer_boundary = plt.Rectangle((0, 0), court_length_ft, court_width_ft, fill="plum", color="plum", lw=2)
    ax.add_patch(outer_boundary)

    # Draw the service lanes with the gap
    service_lane_x1 = 4.5  # the distance between the service lane from the width border line
    service_lane_x2 = service_lane_x1 + service_lane_length_ft
    ax.plot([0, court_length_ft], [service_lane_x1, service_lane_x1], color="black", lw=2)

    # Draw the additional service lines on both sides of the net
    additional_service_line_x1 = 18  # 18 ft from the start (left side of court)
    additional_service_line_x2 = court_length_ft - 18  # Opposite side (right side of court)
    additional_service_line_y1 = 4.5  #height
    additional_service_line_y2 = 31.5  # Adjusted to 27 feet in height
    ax.plot([additional_service_line_x1, additional_service_line_x1], [additional_service_line_y1, additional_service_line_y2], color="black", lw=2)
    ax.plot([additional_service_line_x2, additional_service_line_x2], [additional_service_line_y1, additional_service_line_y2], color="black", lw=2)

    # Draw center service line
    # Measurements are in feets
    center_service_x1 = 18
    center_service_x2 = 60 
    center_service_y1 = 18  # Starting height
    center_service_y2 = 18  # Ending height
    ax.plot([center_service_x1, center_service_x2], [center_service_y1, center_service_y2], color="black", lw=2)

    # Draw the net
    net_x = court_length_ft / 2
    net_y1 = (court_width_ft - net_height_ft) / 2
    net_y2 = (court_width_ft + net_height_ft) / 2
    ax.plot([net_x, net_x], [net_y1, net_y2], color="black", lw=2)

    # Draw the middle line
    middle_x = court_length_ft / 2
    ax.plot([middle_x, middle_x], [0, court_width_ft], color="black", lw=3)

    # Draw the outer lines
    outer_line_y1 = double_alley_width_ft
    outer_line_y2 = court_width_ft - double_alley_width_ft
    ax.plot([0, court_length_ft], [outer_line_y1, outer_line_y1], color="black", lw=2)
    ax.plot([0, court_length_ft], [outer_line_y2, outer_line_y2], color="black", lw=2)

    # Draw the double alleys
    left_double_alley_x = 0
    right_double_alley_x = court_length_ft
    ax.plot([left_double_alley_x, right_double_alley_x], [outer_line_y1, outer_line_y1], color="mediumvioletred", lw=2)
    ax.plot([left_double_alley_x, right_double_alley_x], [outer_line_y2, outer_line_y2], color="mediumvioletred", lw=2)

    # Set the axis ticks and labels
    x_ticks = [0, additional_service_line_x1, middle_line_length_ft, additional_service_line_x2, court_length_ft]
    y_ticks = [0, service_lane_x1, service_lane_x2, court_width_ft]
    ax.set_xticks(x_ticks)
    ax.set_yticks(y_ticks)
    ax.set_xticklabels([f"{val} ft" for val in x_ticks])
    ax.set_yticklabels([f"{val} ft" for val in y_ticks])

    # Save the figure as a PNG file
    plt.savefig(filename, bbox_inches='tight', pad_inches=0, dpi=300)

# Call the function to create and save the tennis court image
draw_tennis_court("tennis_court.png")
```
